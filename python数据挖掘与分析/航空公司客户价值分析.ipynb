## 数据清洗

```python
import pandas as pd
input_file='E://learnFile//chapter7//demo//data/air_data.csv'
data=pd.read_csv(input_file)
data=data[data['SUM_YR_1'].notnull()*data['SUM_YR_2'].notnull()]
index1=data['SUM_YR_1']!=0#index1的结果为true或者false
index2=data['SUM_YR_2']!=0
index3=(data['SEG_KM_SUM']==0)&(data['avg_discount']==0)
```

```python
data=data[index1|index2|index3]
```

## K-Meabs聚类算法

```python
import pandas as pd
from sklearn.cluster import KMeans
input_file='E://learnFile//chapter7//demo//data/zscoreddata.xls'
data=pd.read_excel(input_file)
```

```python
data.head()
```

```python
k=5
model=KMeans(n_clusters=k,n_jobs=4)
model.fit(data)
```

```python
model.cluster_centers_#查看聚类中心
```

```python
model.labels_#查看样本类别
```

```python
data['类别']=model.labels_
```
